```python
Eng = float(input("Enter English Marks: "))
Comp = float(input("Enter Computer Marks: "))
Sci = float(input("Enter Science Marks: "))
Math = float(input("Enter Maths Marks: "))
PakStu = float(input("Enter Pakistan Studies Marks: "))

obtain = Eng + Comp + Sci + Math + PakStu
per = obtain / 500 * 100

if per >= 80 and per <= 100:
    grade = "A"
elif per >= 70 and per < 80:
    grade = "B"
elif per >= 60  and per <= 50:
    grade = "C"
else:
    grade = "Fail"

print(" ")
print("==========================")
print("English Marks(100): ",Eng)
print("Computer Marks(100): ",Comp)
print("General Science Marks(100): ",Sci)
print("Mathematics Marks(100): ",Math)
print("Pakistan Studies Marks(100): ",PakStu)
print(" ")
print("Total Marks Obtained(500): ",obtain)
print("Percentage(%):", per)
print("Grade:", grade)
```
