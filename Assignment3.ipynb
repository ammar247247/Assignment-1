```python
def add(x, y): return x + y
def subtract(x, y): return x - y
def multiply(x, y): return x * y
def divide(x, y): return x / y if y != 0 else "Error"
def power(x, y): return x ** y

def calculator():
    while True:
        print("\n1.Add 2.Sub 3.Mul 4.Div 5.Pow 6.Exit")
        choice = input("Enter choice (1-6): ")
        if choice == '6': break
        if choice in ('1', '2', '3', '4', '5'):
            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
                if choice == '1': print(f"Result: {add(num1, num2)}")
                elif choice == '2': print(f"Result: {subtract(num1, num2)}")
                elif choice == '3': print(f"Result: {multiply(num1, num2)}")
                elif choice == '4': print(f"Result: {divide(num1, num2)}")
                elif choice == '5': print(f"Result: {power(num1, num2)}")
            except ValueError:
                print("Invalid input.")
        else:
            print("Invalid choice")

if __name__ == "__main__":
    calculator()
```

```python
DataStore = [53, "Computer", "LCD", 5.1, 13, "Keyboard"]
Numbers = []

for i in DataStore:
    if isinstance(i, (int, float)):
        Numbers.append(i)
        print(f"{i} is Numeric")
    else:
        print(f"{i} is not Numeric")

print(f"\nFound Numeric Values: {Numbers}")
```

```python
List = [100, 80, 60, 20, 40, 20, 100]

i = set()
Duplicates = []

for a in List:
    if a in i:
        
        if a not in Duplicates:
            Duplicates.append(a)
    else:
        i.add(a)

print(Duplicates)
```
